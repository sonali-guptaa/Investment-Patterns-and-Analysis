# Level 3 Task 5: Reasons for Investment

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importing the dataset
df=pd.read_csv("C:/Users/Rohit Gupta/Downloads/cognify dataset.csv")
```

```python
print(df.columns)
```

## In order to summarize the reasons for the people's investment choices, I decided to analyze the columns Reason_Equity, Reason_Mutual, Reason_Bonds, Reason_FD

```python
# Investigating the columns where participants provided reasons for their investment choices
frequency_equity = df['Reason_Equity'].value_counts()
print("Reasons for choosing equity:\n", frequency_equity)
```

```python
frequency_mutual = df['Reason_Mutual'].value_counts()
print("Reasons for choosing mutual funds:\n", frequency_mutual)
```

```python
frequency_bonds = df['Reason_Bonds'].value_counts()
print("Reasons for choosing government bonds:\n", frequency_bonds)
```

```python
frequency_FD = df['Reason_FD'].value_counts()
print("Reasons for choosing FD:\n", frequency_FD)
```

## After calculating the frequencies for each reason (Equity, Mutual Funds, FD, Bonds), I can combine these results into a single DataFrame for a consolidated view.

```python
# Combining all the frequency data into a single DataFrame
combined_frequencies = pd.concat([frequency_equity, frequency_mutual, frequency_FD, frequency_bonds], axis=1, keys=['Equity', 'Mutual Funds', 'FD', 'Bonds'])

# Summarize the most common reasons
summary = combined_frequencies.sum(axis=1).sort_values(ascending=False)

print("Summary of the most common reasons across all investment types:")
print(summary)
```

```python
# Visualizing the summarizing data
combined_frequencies.plot(kind='bar', figsize=(10, 6))
plt.title('Frequency of Reasons for Different Investment Types')
plt.xlabel('Reasons')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

## So, based on the above calculation and bar chart, top Reasons for Investment Choices includes:
1)Capital Appreciation (30):
This is the most frequently cited reason for investments, indicating that a significant number of participants are focused on the potential for their investments to increase in value over time.

2)Assured Returns (26):
The second most common reason, showing that many investors are looking for stable and predictable returns.

3)Better Returns (24):
Close behind, this reason highlights the participants' desire for investments that outperform average returns.

4)Risk-Free (19):
This indicates a strong preference for investments that are perceived as having minimal or no risk.

5)Fixed Returns (18):
Similar to assured returns, this reason suggests that many participants are attracted to investments that provide fixed returns.

## So, overall the data shows that most participants are primarily motivated by the potential for capital appreciation, assured and better returns, and risk aversion. Fixed returns and diversification are also significant factors. This suggests that the majority of investors are looking for a balance between growth and safety in their investment choices.

# Level 3 Task 6: Savings Objectives

```python
# Analyzing the saving objectives
savings_objectives = df['What are your savings objectives?'].value_counts()
print("List of Savings Objectives and Their Frequencies: \n",savings_objectives)
```

```python
# Plotting a bar chart for the savings objectives
plt.figure(figsize=(10, 6))
savings_objectives.plot(kind='bar', color='blue')
plt.title('Savings Objectives Frequency')
plt.xlabel('Savings Objectives')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

## Retirement Plan is the most common savings objective with 24 participants mentioning it and health Care is the second most common, with 13 participants prioritizing this objective. Also, education is the least common savings objective with 3 participants only.

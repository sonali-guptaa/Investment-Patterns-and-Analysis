# Level 4 Task 7: Common Information Sources

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing the dataset
df=pd.read_csv("C:/Users/Rohit Gupta/Downloads/cognify dataset.csv")
```

```python
print(df.columns)
```

```python
# Reviewing the source column where participants indicated their sources of investment information
information_sources = df['Source'].value_counts()
print("List of Information Sources and Their Frequencies:\n", information_sources)
```

```python
# Plotting a bar chart for the information sources
plt.figure(figsize=(10, 6))
information_sources.plot(kind='bar', color='green')
plt.title('Information Sources Frequency')
plt.xlabel('Information Sources')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()
```

## So, as per the above bar chart it is clear that the most common source that participants rely on is Financial Consultants. Also, the second most common source is newspapers and magazines.

# Level 4 Task 8: Investment Duration

```python
print(df['Duration'])
```

## As the Duration column contains categorical values representing different ranges, we can map these categories to numeric values to calculate the average duration.

```python
# Mapping for the duration categories to numeric values
duration_mapping = {
    'Less than 1 year': 0.5,
    '1-3 years': 2,
    '3-5 years': 4,
    'More than 5 years': 6
}
```

```python
# Converting 'Duration' to numeric values based on the mapping
df['Duration_numeric'] = df['Duration'].map(duration_mapping)
```

```python
# Calculating the average investment duration
average_duration = df['Duration_numeric'].mean()
print(f"The average investment duration is: {average_duration:} years")
```

## So, the average investment duration is 2.9 years.

# Level 4 Task 10:Correlation Analysis

```python
# Mapping for the 'Duration' column
duration_mapping = {
    'Less than 1 year': 0.5,
    '1-3 years': 2,
    '3-5 years': 4,
    'More than 5 years': 6
}

# Map 'Duration' to numeric values
df['Duration_numeric'] = df['Duration'].map(duration_mapping)

# Mapping for the 'Expect' column
expect_mapping = {
    '10%-20%': 15,
    '20%-30%': 25,
    '30%-40%': 35
}

# Map 'Expect' to numeric values
df['Expect_numeric'] = df['Expect'].map(expect_mapping)

# Select relevant columns
relevant_columns = ['Age', 'Duration_numeric', 'Expect_numeric']
df_relevant = df[relevant_columns]
```

```python
# Calculating correlation matrix
correlation_matrix = df_relevant.corr()
print("Correlation Matrix: \n",correlation_matrix)

# Plotting the correlation matrix as a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

## From the correlation matrix shown in the heatmap:

1) The correlation between Age and Duration_numeric is very weak (0.05). This suggests that there's almost no linear relationship between age and the investment duration.
2) The correlation between Age and Expect_numeric is -0.09, which is also very weak and slightly negative. This indicates that as age increases, the expected return percentage might decrease slightly, but the relationship is very weak.
3) The correlation of Duration_numeric with Expect_numeric is 0.25, indicating a weak positive correlation. This suggests that as the investment duration increases, the expected return percentage tends to increase slightly.
4) Expect_numeric has a weak negative correlation with Age and a weak positive correlation with Duration_numeric.

```python
# Scatter plots for Age vs Duration
plt.figure(figsize=(10, 5))
plt.scatter(df_relevant['Age'], df_relevant['Duration_numeric'], alpha=0.7)
plt.title('Age vs Investment Duration')
plt.xlabel('Age')
plt.ylabel('Investment Duration (Years)')
plt.show()
```

## The scatterplot supports the idea that the age of an individual does not strongly determine their chosen investment duration. There are clusters of preferred investment durations, but they are spread across different ages, indicating weak or no linear correlation between the two variables.

```python
# Scatter plots for Age vs Expectation
plt.figure(figsize=(10, 5))
plt.scatter(df_relevant['Age'], df_relevant['Expect_numeric'], alpha=0.7)
plt.title('Age vs Expected Returns')
plt.xlabel('Age')
plt.ylabel('Expected Returns (%)')
plt.show()
```

## The scatterplot suggests that expected returns are fairly consistent across different age groups, with no significant trend that indicates a strong relationship between age and expected return percentages. This reinforces the conclusion that age is not a strong determinant of expected investment returns in this dataset.

```python
# Scatter plots for Duration vs Expectation
plt.figure(figsize=(10, 5))
plt.scatter(df_relevant['Duration_numeric'], df_relevant['Expect_numeric'], alpha=0.7)
plt.title('Duration vs Expectation Returns')
plt.xlabel('Investment Duration')
plt.ylabel('Expected Returns (%)')
plt.show()
```

## As per the above scatter plot, there seems to be a weak positive correlation exists between Investment Duration and Expected Returns. This suggests that, on average, longer-duration investments are associated with higher expected returns.

## So, overall, the variables show very weak correlations with one another in the above correlation analysis. This suggests that there isn't a strong linear relationship between these variables, meaning that changes in one variable are not strongly associated with changes in the others. This implies that other factors may be influencing these variables, and further investigation would be needed to identify those factors.

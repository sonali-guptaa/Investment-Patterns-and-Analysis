# Level 2 Task 4: Most Preferred Investment Avenuee

```python
print(df.columns)
```

```python
# Frequency Analysis
frequency_analysis = df['Avenue'].value_counts()
print("Frequency Analysis:\n", frequency_analysis)
```

```python
# visual representation of frequency analysis
plt.figure(figsize=(8, 6))
frequency_analysis.plot(kind='bar', color=['blue', 'green', 'yellow','red'])
plt.title('Investment Avenue Frequency Analysis')
plt.xlabel('Investment Avenue Types')
plt.ylabel('Frequency Count')
plt.xticks(rotation=0)
plt.show()
```

## So, it's clear from the above bar chart that Mutual Fund is the most preferred investment avenue with the highest frequency.
